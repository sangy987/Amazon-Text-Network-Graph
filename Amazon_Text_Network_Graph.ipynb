```python
words = open("/content/clean_amazon_file.txt",'r').read().splitlines()
```

```python
import pandas as pd
import networkx
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter
from bokeh.io import output_notebook, show, save
output_notebook()
from bokeh.io import output_notebook, show, save
from bokeh.models import Range1d, Circle, ColumnDataSource, MultiLine
from bokeh.plotting import figure
from bokeh.plotting import from_networkx
from bokeh.palettes import Turbo256
from bokeh.transform import linear_cmap
```

```python
import networkx as nx
amazon = nx.Graph()
```

### The 100 most frequently used words are added as nodes to the graph, with the weights being the frequency of the words appearing in the text.

```python
Counter = Counter(words)
nodes = Counter.most_common(100)
# print(nodes)
```

### Created bigrams and added the first 1000 most common bigrams as edges. Similarly, the weights were the frequency of the bigrams.

```python
from collections import Counter
bigrams = zip(words, words[1:])
Counter= Counter(bigrams)
bigram =(Counter.most_common(1000))
# print(bigram)
```

```python
#nodes in the graph
for i in nodes:
  amazon.add_node(i[0],weight=i[1])
amazon.nodes
```

```python
#edges in the graph
for i in bigram:  
  amazon.add_edge(i[0][0],i[0][1], weight=i[1])
amazon.edges()
```

```python
pos = nx.spring_layout(amazon)
print(pos)
```

```python
#calculating the degree of each node
degrees = dict(networkx.degree(amazon))
networkx.set_node_attributes(amazon, name='degree', values=degrees)
```

```python
#adjusted the size so that even less weighted nodes are visible
number_to_adjust_by = 5
adjusted_node_size = dict([(node, degree+number_to_adjust_by) for node, degree in networkx.degree(amazon)])
print(adjusted_node_size)
networkx.set_node_attributes(amazon, name='adjusted_node_size', values=adjusted_node_size)
```

### Used the bokeh palette Turbo256 to colour the nodes because of large number of nodes

```python
from bokeh.models import EdgesAndLinkedNodes, NodesAndLinkedEdges,EdgesAndLinkedNodes, NodesAndLinkedEdges, LabelSet
node_highlight_color = 'white'
edge_highlight_color = 'black'

size_by_this_attribute = 'adjusted_node_size'

title = 'Amazon Text R Network Graph'

HOVER_TOOLTIPS = [
       ("Words", "@index"),
        ("Degree", "@degree"),
         
]
#made the plot responsive and added the xrange and yrange along with the width and height
plot = figure(tooltips = HOVER_TOOLTIPS,
              tools="pan,wheel_zoom,save,reset", active_scroll='wheel_zoom',
            x_range=Range1d(-1, 1), y_range=Range1d(-1, 1), title=title, width=1000,height=1000,sizing_mode='scale_width' )

network_graph = from_networkx(amazon, networkx.spring_layout, scale=10, center=(0, 0))
network_graph.node_renderer.data_source.data['colors'] = Turbo256
network_graph.node_renderer.glyph = Circle(size=size_by_this_attribute, fill_color='colors')
network_graph.node_renderer.hover_glyph = Circle(size=size_by_this_attribute, fill_color=node_highlight_color, line_width=2)
network_graph.node_renderer.selection_glyph = Circle(size=size_by_this_attribute, fill_color=node_highlight_color, line_width=2)

network_graph.edge_renderer.glyph = MultiLine(line_alpha=0.5, line_width=1)
network_graph.edge_renderer.selection_glyph = MultiLine(line_color=edge_highlight_color, line_width=2)
network_graph.edge_renderer.hover_glyph = MultiLine(line_color=edge_highlight_color, line_width=2)

network_graph.selection_policy = NodesAndLinkedEdges()
network_graph.inspection_policy = NodesAndLinkedEdges()

network_graph.edge_renderer.glyph = MultiLine(line_alpha=0.5, line_width=1)
x, y = zip(*network_graph.layout_provider.graph_layout.values())
node_labels = list(amazon.nodes())
#added labels of the nodes
source = ColumnDataSource({'x': x, 'y': y, 'name': [node_labels[i] for i in range(len(x))]})
labels = LabelSet(x='x', y='y', text='name', source=source, background_fill_color='white', text_font_size='15px', background_fill_alpha=.7)
plot.renderers.append(labels)
plot.renderers.append(network_graph)

show(plot)
```
